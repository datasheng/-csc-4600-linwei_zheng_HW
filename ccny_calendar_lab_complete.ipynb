# CCNY Academic Calendar Web Scraper

This notebook scrapes the **CCNY Fall 2021 Academic Calendar** using:
- requests
- BeautifulSoup
- pandas

Final result: a pandas DataFrame where:
- The index is a Python date
- Column `dow` = day of the week
- Column `text` = event description


## Step 1 — Import Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Step 2 — Download the Webpage

```python
url = 'https://www.ccny.cuny.edu/registrar/fall-2021-academic-calendar'
response = requests.get(url)
html = response.text
print('Status Code:', response.status_code)
```

## Step 3 — Parse HTML

```python
soup = BeautifulSoup(html, 'html.parser')
tables = soup.find_all('table')
print('Tables found:', len(tables))
```

## Step 4 — Extract Table Rows

```python
rows = tables[0].find_all('tr')
data = []

for row in rows[1:]:
    cols = row.find_all('td')
    if len(cols) == 3:
        date = cols[0].get_text(strip=True)
        dow = cols[1].get_text(strip=True)
        text = cols[2].get_text(' ', strip=True)
        data.append([date, dow, text])
```

## Step 5 — Create Pandas DataFrame

```python
df = pd.DataFrame(data, columns=['date', 'dow', 'text'])
df.head()
```

## Step 6 — Convert Date Column

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

## Step 7 — Set Date as Index

```python
df = df.set_index('date')
df
```
